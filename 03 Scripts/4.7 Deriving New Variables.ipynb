# If-Statements with User-Defined Functions

```python
#Creating path string
path = r'C:\Users\Alex Lam\Desktop\Career Foundary\04 Instacart Project'
```

```python
#Importing prior orders dataframe
df_ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
#Testing dataframe output
df_ords_prods_merged.head()
```

```python
#Cleaning up dataframe
df_ords_prods_merged_cleaned = df_ords_prods_merged.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', '_merge'])
```

```python
#New subset to create new column
df = df_ords_prods_merged_cleaned[:1000000]
```

```python
#Define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
#There is no product in the 'High Range'
df['prices'].max()
```

```python
df.head()
```

# If-Statements with the loc() Function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
bar_price_range = df['price_range_loc'].value_counts().plot.bar()
plt.title('Price Range Count')
plt.xlabel('Price Range')
plt.ylabel('Price Range Count')
plt.savefig(os.path.join(path, '04 Analysis','Visualizations', 'Price Range Count'), bbox_inches='tight')
```

# Busiest Day Column for Ords_Prods_Merged

# If-Statements with For-Loops

```python
#Seeing which days are the busiest
df_ords_prods_merged_cleaned['orders_day_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in df_ords_prods_merged_cleaned["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
df_ords_prods_merged_cleaned['busiest_day'] = result
```

```python
#Testing dataframe output after new column
df_ords_prods_merged_cleaned.head()
```

```python
df_ords_prods_merged_cleaned['busiest_day'].value_counts(dropna = False)
```

# Creating Busiest Days and Slowest Days

```python
#2.
#Using for loop along with or operator to state which days are busiest, least busiest, or have regular business
#Using days as new list
days = []

for x in df_ords_prods_merged_cleaned["orders_day_of_week"]:
  if (x == 0) or (x == 1):
    days.append("Busiest days")
  elif (x == 3) or (x == 4):
    days.append("Least busiest days")
  else:
    days.append("Regularly busy")
```

```python
#3.
#Testing days list
days
```

```python
#Creating new busiest_days column
df_ords_prods_merged_cleaned['busiest_days'] = days
```

```python
#Testing dataframe output after new column
df_ords_prods_merged_cleaned.head()
```

```python
#Comparing value count to busiest day column, previous column created
#Busiest and least busiest days have increased while regularly busy has decreased, to be expected
df_ords_prods_merged_cleaned['busiest_days'].value_counts(dropna = False)
```

```python
Average = [17, 8, 18, 19, 20, 7, 21, 22]
```

```python
#Using for loop assign strings to order_hour_of_days values
#Fewest orders is being used in the else condition
hours = []

for x in df_ords_prods_merged_cleaned["order_hour_of_day"]:
  if (x in Most):
    hours.append("Most orders")
  elif (x in Average):
    hours.append("Average orders")
  else:
    hours.append("Fewest orders")
```
