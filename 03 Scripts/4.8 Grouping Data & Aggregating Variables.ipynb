```python
#4.
#Creating loyalty flag existing for customers
df_ords_prods_merged_clean['max_order'] = df_ords_prods_merged_clean.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Creating loyalty column for customers based on max_order (loyalty flag) value
df_ords_prods_merged_clean.loc[df_ords_prods_merged_clean['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merged_clean.loc[(df_ords_prods_merged_clean['max_order'] <= 40) & (df_ords_prods_merged_clean['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merged_clean.loc[df_ords_prods_merged_clean['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

# df_loyality_flag_avg

```python
#5.
#Means for each loyalty flag
df_loyality_flag_avg = df_ords_prods_merged_clean.groupby('loyalty_flag').agg({'prices': ['mean']})
```

# df_spender_avg

```python
#6.
#New dataframe for prices mean
df_spender_avg = df_ords_prods_merged_clean.groupby('user_id').agg({'prices': ['mean']})
```

```python
df_spender_avg.head()
```

```python
#Creating price_avg column for loc function
df_spender_avg['price_avg'] = df_ords_prods_merged_clean.groupby('user_id').agg({'prices': ['mean']})
```

```python
#loc function to determine which users are high or lower spenders
df_spender_avg.loc[df_spender_avg['price_avg'] >= 10, 'spending_avg'] = 'High spender'
df_spender_avg.loc[df_spender_avg['price_avg'] < 10, 'spending_avg'] = 'Low spender' 
```

```python
#Checking dataframe after loc function
df_spender_avg.head()
```

```python
#Spender Distribution
df_spender_avg['spending_avg'].value_counts(dropna = False)
```

# df_customer_frequency

```python
#7.
#New dataframe for days_since_prior_order mean
df_customer_frequency = df_ords_prods_merged_clean.groupby('user_id').agg({'days_since_prior_order': ['mean']})
```

```python
#Creating price_avg column for loc function
df_customer_frequency['frequency_mean'] = df_ords_prods_merged_clean.groupby('user_id').agg({'days_since_prior_order': ['mean']})
```

```python
#loc function to determine how frequent users shop
df_customer_frequency.loc[df_customer_frequency['frequency_mean'] > 20, 'customer_frequency'] = 'Non-frequent customer'
df_customer_frequency.loc[(df_customer_frequency['frequency_mean'] <= 20) & (df_customer_frequency['frequency_mean'] > 10), 'customer_frequency'] = 'Regular customer'
df_customer_frequency.loc[df_customer_frequency['frequency_mean'] <= 10, 'customer_frequency'] = 'Frequent customer'
```

```python
df_customer_frequency.head()
```

```python
#Frequency Distribution
df_customer_frequency['customer_frequency'].value_counts(dropna = False)
```

# df_user_stats

```python
#8.
#Creating new dataframe with all user_id stats
df_user_stats = df_spender_avg.merge(df_customer_frequency, on = ['user_id'], indicator = True) 
```

```python
df_user_stats.head()
```

```python
#Cleaning repetitive columns
df_user_stats_cleaned = df_user_stats.drop(columns = ['prices', 'days_since_prior_order', '_merge'])
```
